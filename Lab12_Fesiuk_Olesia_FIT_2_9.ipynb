```python
#  11
#:  
#  
# 2-9
# 28
#https://github.com/Olllesssyya/Lab12_numerical

#   ' . 
import numpy as np
import matplotlib.pyplot as plt 

def f(x, y):
    return x+np.cos(y/np.sqrt(1.3))

a = 1.2 #   
b = 2.2 #   
h = 0.1 # 
y0 = 1.8 #  

n = int((b - a) / h) #  

x = np.array ([i*h for i in range(0, 11)]) #  x  

y=np.empty(n+1) 
y[0] = y0

for i in range (n):
    y[i+1]=y[i]+f(x[i],y[i])*h

y_rounded=np.round_(y,4) 
print("x=",x, "\ny=",y_rounded) 

plt.plot(x,y, "o",x,y, "red") 
plt.xlabel("x") 
plt.ylabel("y") 
plt.title(" ") 
plt.legend(["", "sin(x)-cos(y)"]) 
plt.grid() 
plt.show()

# -
def f(x, y):
    return np.sin(x) - np.cos(y)

a = 0.5 #   
b = 1.5 #   
h = 0.1 # 
y0 = 1.8 #  
n = int((b - a) / h) #  
x = np.array ([i*h for i in range(0, 11 )]) #  x  

y=np.empty(n+1) 
y[0] = y0
for i in range(n): 

     y[i+1]=y[i]+(f(x[i],y[i])+f(x[i+1],y[i])+h*f(x[i],y[i]))*h/2 

y_rounded=np.round_(y,4)

print("x=",x, "\ny=",y_rounded) 
plt.plot(x,y, "o",x,y, "red") 
plt.xlabel("x") 
plt.ylabel("y") 
plt.title(" -") 
plt.legend(["", "sin(x)-sin(y)"]) 
plt.grid() 
plt.show()


```
